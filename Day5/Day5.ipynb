```python
with open("Day5_input.txt", 'r') as input_file:
    input_raw = input_file.read().split('\n\n')

ranges_raw = input_raw[0]
ids_raw = input_raw[1]
```

```python
# IDs processing

ids = list(map(int, ids_raw.split('\n')))
print(type(ids[0]))
```

```python
# Ranges processing

id_ranges = []
id_ranges_raw = ranges_raw.split('\n')

for id_range_raw in id_ranges_raw:
    int_range = list(map(int, id_range_raw.split('-')))
    id_ranges.append(int_range)
```

```python
# Checking IDs for freshness
fresh_count = 0

for id in ids:
    for id_range in id_ranges:
        if id >= id_range[0] and id <= id_range[1]:
            fresh_count += 1
            print(f'Fresh ingredient with ID:{id}')
            break

print(f'Found {fresh_count} fresh ingredients')
```

```python
def merge_ranges(id_ranges):
    # Merge overlapping ranges

    merged_id_ranges = []

    # Overlap counter
    overlaps = 0

    for i in range(len(id_ranges)-1):
        this_id_range = id_ranges[i]
        for other_id_range in id_ranges[(i+1):]:
            # Enclosed - delete range
            if this_id_range[0] >= other_id_range[0] and this_id_range[1] <= other_id_range[1]:
                overlaps += 1
                print(f'(All inside) Range {this_id_range} is inside {other_id_range}')
                continue
            # Beginning inside other range (merge other beginning to this end)
            if this_id_range[0] >= other_id_range[0] and this_id_range[0] <= other_id_range[1]:
                overlaps += 1
                merged_range = [other_id_range[0], this_id_range[1]]
                merged_id_ranges.append(merged_range)
                print(f'(Beginning inside) Merged {this_id_range} and {other_id_range} into {merged_range}')
            # End inside other range (merge this beginning to other end)
            if this_id_range[1] >= other_id_range[0] and this_id_range[1] <= other_id_range[1]:
                overlaps += 1
                merged_range = [this_id_range[0], other_id_range[1]]
                merged_id_ranges.append(merged_range)
                print(f'(End inside) Merged {this_id_range} and {other_id_range} into {merged_range}')

    print(f'Found {overlaps} overlaps')

    return merged_id_ranges, overlaps
```
