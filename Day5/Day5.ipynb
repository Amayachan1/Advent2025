```python
# IDs processing

ids = list(map(int, ids_raw.split('\n')))
print(type(ids[0]))
```

```python
# Ranges processing

id_ranges = []
id_ranges_raw = ranges_raw.split('\n')

for id_range_raw in id_ranges_raw:
    int_range = list(map(int, id_range_raw.split('-')))
    int_proper_range = range(int_range[0], int_range[1]+1)
    id_ranges.append(int_proper_range)

sorted_id_ranges = sorted(id_ranges, key=lambda r: r.start)
```

```python
# Checking IDs for freshness
fresh_count = 0

for id in ids:
    for id_range in sorted_id_ranges:
        if id in id_range:
            fresh_count += 1
            print(f'Fresh ingredient with ID:{id}')
            break

print(f'Found {fresh_count} fresh ingredients')
```

```python
def ranges_overlap(range1, range2):
    if max(range1.start, range2.start) <= min(range1.stop, range2.stop):
        return 1
    else:
        return 0
```

```python
def merge_ranges(range1, range2):
    return range(min(range1.start, range2.start), max(range1.stop, range2.stop))
```

```python
def find_non_unique(id_ranges):

    non_unique_ranges_index = []

    for i, id_range in enumerate(id_ranges[:-1]):
        for j, other_id_range in enumerate(id_ranges[i+1:]):
            if ranges_overlap(id_range, other_id_range) == 1:
                non_unique_ranges_index.append(i)
                non_unique_ranges_index.append(i+1+j)
                # print(f'Overlapping ranges {id_range} and {other_id_range}')
    
    return set(non_unique_ranges_index)
```

```python
def merge_all_ranges(id_ranges):

    merged_id_ranges = []

    # Find unique and non unique indexes
    non_unique_indexes = find_non_unique(id_ranges)
    unique_indexes = [i for i in range(len(id_ranges)) if i not in non_unique_indexes]

    # Add unique ranges
    for index in unique_indexes:
        merged_id_ranges.append(id_ranges[index])

    # Make non uniqe ranges list
    non_unique_ranges = []
    for index in non_unique_indexes:
        non_unique_ranges.append(id_ranges[index])

    # Overlap counter
    overlaps = 0

    for i, id_range in enumerate(non_unique_ranges[:-1]):
        for other_id_range in non_unique_ranges[i+1:]:
            # print(f'Comparing ranges {id_range} and {other_id_range}')

            if ranges_overlap(id_range, other_id_range) == 1:
                overlaps += 1
                merged = merge_ranges(id_range, other_id_range)
                merged_id_ranges.append(merged)
                break
                # print(f'Merged {id_range} and {other_id_range} into {merged}')
    
    print(f'Found {overlaps} overlaps')
    
    return merged_id_ranges, overlaps
            
```

```python
with open("Day5_input.txt", 'r') as input_file:
    input_raw = input_file.read().split('\n\n')

ranges_raw = input_raw[0]
ids_raw = input_raw[1]

# Ranges processing

id_ranges = []
id_ranges_raw = ranges_raw.split('\n')

for id_range_raw in id_ranges_raw:
    int_range = list(map(int, id_range_raw.split('-')))
    int_proper_range = range(int_range[0], int_range[1]+1)
    id_ranges.append(int_proper_range)

sorted_id_ranges = sorted(id_ranges, key=lambda r: r.start)
```

```python
overlaps = -1

while overlaps != 0:
    sorted_id_ranges, overlaps =  merge_all_ranges(sorted_id_ranges)

print(sorted_id_ranges)
```
