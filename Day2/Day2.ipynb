```python
with open("Day2_input.txt", 'r') as input_file:
    ranges_raw = input_file.read().split(',')
```

Part 1

```python
sum = 0
for range_raw in ranges_raw:
    range_nums = [int(i) for i in range_raw.split('-')]
    
    for id in range(range_nums[0], range_nums[1]+1):
        id_string = str(id)
        center = int(len(id_string)/2)
        if id_string[:center] == id_string[center:]:
            sum += id

print(sum)
```

Part 2

```python
invalid_ids = []
for range_raw in ranges_raw:
    range_nums = [int(i) for i in range_raw.split('-')]
    
    for id in range(range_nums[0], range_nums[1]+1):
        id_string = str(id)
        id_length = len(id_string)
        center = int(id_length/2)
        for i in range(1, center+1):
            # check that it splits evenly
            if id_length % i == 0:
                unit_count = int(id_length / i)
                unit = id_string[:i]
                repeat_id = unit * unit_count
                if id == int(repeat_id):
                    if id not in invalid_ids:
                        invalid_ids.append(id)

sum = 0
for id in invalid_ids:
    sum += id

print(sum)
```
