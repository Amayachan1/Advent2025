```python
zero_count = 0
marker = 50
```

```python
with open("Day1_input.txt", 'r') as input_file:
    instructions = [line[:-1] for line in input_file]
```

```python
for move in instructions:
    # split input
    direction = move[0]
    steps = int(move[1:]) % 100
    #print(f'Marker:{marker}')
    old_marker = marker
    # move the dial
    if direction == 'R':
        marker += steps
    else:
        marker -= steps
    # check for going over the 0/99
    if marker > 99:
        marker -= 100
    if marker < 0:
        marker += 100
    # check for 0s:
    if marker == 0:
        zero_count += 1
    print(f'Old:{old_marker}, Instruction:{move}, New marker:{marker}, Zero count:{zero_count}')
    
```

```python
print(zero_count)
```

```python
# Reset the dial
zero_count = 0
marker = 50
```

```python
for move in instructions:
    # split input
    direction = move[0]
    steps = int(move[1:]) % 100
    # Add any 0 passes
    zero_count += (int(move[1:]) // 100)
    old_marker = marker
    if steps == 0:
        continue
    # move the dial
    if direction == 'R':
        marker += steps
        if marker > 99:
            marker -= 100
            if marker != 0:
                zero_count += 1
    else:
        marker -= steps
        if marker < 0:
            marker += 100
            if old_marker != 0:
                zero_count += 1
 # check for 0s:
    if marker == 0:
        zero_count += 1
    print(f'Old:{old_marker}, Instruction:{move}, New marker:{marker}, Zero count:{zero_count}')
```
