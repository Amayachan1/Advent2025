```python
with open("Day6_input.txt", 'r') as input_file:
    input_raw = input_file.read().split('\n')

input_stripped = []
for line in input_raw:
    stripped_line = list(filter(lambda x: x != '', line.split(' ')))
    input_stripped.append(stripped_line)
```

```python
# Format numbers

number_table_raw = input_stripped[:-1]
number_table = []

for row in number_table_raw:
    new_row = [int(num) for num in row]
    number_table.append(new_row)
```

```python
row_length = len(input_stripped[0])
operators = input_stripped[-1]


total_sum = 0

for col_i in range(row_length):
    if operators[col_i] == '+':
        col_result = 0
        for row in number_table:
            col_result += row[col_i]
        print(f'Column {col_i} result is {col_result}')
    if operators[col_i] == '*':
        col_result = 1
        for row in number_table:
            col_result *= row[col_i]
        print(f'Column {col_i} result is {col_result}')

    total_sum += col_result

print(f'Total is {total_sum}')
    
```

```python
with open("Day6_input.txt", 'r') as input_file:
    input_raw = input_file.read().split('\n')
```

```python
operators = list(filter(lambda x: x != '', input_raw[-1].split(' ')))
```

```python
number_table = input_raw[:-1]
row_char_length = len(number_table[0])
col_length = len(number_table)
skip_number = [' ' for i in range(col_length)]
operator_i = 0
numbers_in_operation = []
total_sum = 0

for char_i in range(row_char_length+1):

    operator = operators[operator_i]

    try:
        number = [row[char_i] for row in number_table]
    except IndexError:
        print('Last number')
        number = []
    
    if number == skip_number or char_i == row_char_length:

        if operator == '+':
            operation_result = 0
            for op_num in numbers_in_operation:
                operation_result += op_num
            total_sum += operation_result
            print(f'Column {operator_i} with operator {operator}, result is {operation_result}')
     
        if operator == '*':
            operation_result = 1
            for op_num in numbers_in_operation:
                operation_result *= op_num
            total_sum += operation_result
            print(f'Column {operator_i} with operator {operator}, result is {operation_result}')

        print('NEXT')
        numbers_in_operation = []
        operator_i += 1

    else:
        stripped_num = list(filter(lambda x: x != ' ', number))
        int_number = int(''.join(stripped_num))
        numbers_in_operation.append(int_number)
        print(numbers_in_operation)
```

```python
print(f'Total is {total_sum}')
```
