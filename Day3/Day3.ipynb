```python
banks = []
with open("Day3_input.txt", 'r') as input_file:
    for line in input_file.read().split("\n"):
        bank = [int(battery) for battery in line]
        banks.append(bank)
```

Part 1

```python
max_joltage = 0

for bank in banks:
    max_bank_joltage = 0
    max_single_joltage = max(bank[:-1])
    bank_string = ''.join([str(battery) for battery in bank])

    for first_position in range(len(bank)-1):
        bank_joltage = 0
        first_battery = bank[first_position]
        if first_battery == max_single_joltage:
            rest_of_bank = bank[first_position+1:]
            second_battery = max(rest_of_bank)
            bank_joltage = first_battery*10 + second_battery
            max_bank_joltage = bank_joltage
            break
    
    print(f'Bank {bank_string} \n Max joltage {max_bank_joltage}')
    max_joltage += max_bank_joltage
```

```python
print(max_joltage)
```

Part 2 (solved with recursion)

```python
def find_max_bank_joltage(bank, on_batteries_count):
    # bank_string = ''.join([str(battery) for battery in bank])
    # print(bank_string)

    if on_batteries_count == 0:
        return 0
    
    else:
        slots_to_leave = on_batteries_count - 1
        if slots_to_leave == 0:
            avaliable_bank_for_position = bank
        else:
            avaliable_bank_for_position = bank[:-slots_to_leave]
        max_position_joltage = max(avaliable_bank_for_position)

        for position_option in range(len(avaliable_bank_for_position)):
            
            position_battery = bank[position_option]
            if position_battery == max_position_joltage:
                added_voltage = position_battery*(10**slots_to_leave)
                rest_of_bank = bank[position_option+1:]
                return added_voltage+find_max_bank_joltage(rest_of_bank, on_batteries_count-1)
```

```python
max_joltage = 0

for bank in banks:
    bank_joltage = find_max_bank_joltage(bank, 12)
    max_joltage += bank_joltage

print(max_joltage)
```
